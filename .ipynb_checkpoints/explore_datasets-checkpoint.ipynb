```python
#check how many nan values are in a col
def check_na(col_name,df):
    return sum(df[col_name].isna())


#turn dollars(prices) to 
def numericalize(string):
    if type(string) != str:
        pass
    else:
        num = ''
        for char in list(string):
            if char.isnumeric() or char == '.':
                num += char
        return float(num)
```
